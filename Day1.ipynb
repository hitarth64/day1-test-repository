```python
import numpy as np
import matplotlib.pyplot as plt

# Generate samples
samples = np.random.normal(0, 1, 1000)

# Plot histogram
plt.hist(samples, bins=200)
plt.xlabel('Sample Value')
plt.ylabel('Frequency')
plt.title('Histogram of Normal Distribution Samples')
plt.show()
```
